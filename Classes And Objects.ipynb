```python
#calculate average rating...
class OnlineFoodDelivery:
    def __init__(self, ID, Delivery_person_ID, Delivery_person_Age, Delivery_person_Ratings,
                 Restaurant_latitude, Restaurant_longitude, Delivery_location_latitude,
                 Delivery_location_longitude, Type_of_order, Type_of_vehicle, Time_taken_min):
        self.ID = ID
        self.Delivery_person_ID = Delivery_person_ID
        self.Delivery_person_Age = Delivery_person_Age
        self.Delivery_person_Ratings = Delivery_person_Ratings
        self.Restaurant_latitude = Restaurant_latitude
        self.Restaurant_longitude = Restaurant_longitude
        self.Delivery_location_latitude = Delivery_location_latitude
        self.Delivery_location_longitude = Delivery_location_longitude
        self.Type_of_order = Type_of_order
        self.Type_of_vehicle = Type_of_vehicle
        self.Time_taken_min = Time_taken_min

    def display_delivery_details(self):
        print(f"Delivery ID: {self.ID}")
        print(f"Delivery Person ID: {self.Delivery_person_ID}")
        print(f"Delivery Person Age: {self.Delivery_person_Age}")
        print(f"Delivery Person Ratings: {self.Delivery_person_Ratings}")
        print(f"Restaurant Latitude: {self.Restaurant_latitude}")
        print(f"Restaurant Longitude: {self.Restaurant_longitude}")
        print(f"Delivery Location Latitude: {self.Delivery_location_latitude}")
        print(f"Delivery Location Longitude: {self.Delivery_location_longitude}")
        print(f"Type of Order: {self.Type_of_order}")
        print(f"Type of Vehicle: {self.Type_of_vehicle}")
        print(f"Time Taken (min): {self.Time_taken_min}")

    @staticmethod
    def calculate_average_rating(deliveries):
        total_ratings = sum(delivery.Delivery_person_Ratings for delivery in deliveries)
        average_rating = total_ratings / len(deliveries)
        return average_rating

# Example usage:
delivery1 = OnlineFoodDelivery(ID=1, Delivery_person_ID=101, Delivery_person_Age=30,
                               Delivery_person_Ratings=4.5, Restaurant_latitude=12.971598,
                               Restaurant_longitude=77.594562, Delivery_location_latitude=12.918030,
                               Delivery_location_longitude=77.636811, Type_of_order='Online',
                               Type_of_vehicle='Bike', Time_taken_min=25)

delivery2 = OnlineFoodDelivery(ID=2, Delivery_person_ID=102, Delivery_person_Age=28,
                               Delivery_person_Ratings=4.8, Restaurant_latitude=12.926034,
                               Restaurant_longitude=77.614584, Delivery_location_latitude=12.908045,
                               Delivery_location_longitude=77.647191, Type_of_order='Offline',
                               Type_of_vehicle='Car', Time_taken_min=30)

deliveries = [delivery1, delivery2]

# Display details for each delivery
for delivery in deliveries:
    delivery.display_delivery_details()
    print("\n")

# Calculate and display the average rating
average_rating = OnlineFoodDelivery.calculate_average_rating(deliveries)
print(f"Average Rating for Delivery Persons: {average_rating}")
```
